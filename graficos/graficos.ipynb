```python
import matplotlib.pyplot as plt
```

# Bloco 1

## Gráfico 1
Contexto/Ambiente dos afetos dos empréstimos africanos encontrados no crioulo guineense

```python
title1 = "Contexto/Ambiente dos afetos dos empréstimos africanos encontrados no crioulo guineense"
collumn_type = [
    "Familiar",
    "Religioso",
    "Comemoração",
    "Amizade",
    "Acolhimento",
    "Nomeação",
    "Indignação", 
    "Amoroso",
    "Animosidade"
]
collumn_values = [7, 3, 1, 2, 1, 3, 1, 1, 1]
```

```python
# Create the histogram (using a sorted bar chart for better visualization of frequency)
sorted_data = sorted(zip(collumn_values, collumn_type), reverse=True)
sorted_values = [x[0] for x in sorted_data]
sorted_labels = [x[1] for x in sorted_data]

plt.figure(figsize=(10, 6))
plt.bar(x=sorted_labels, height=sorted_values, color='purple')
plt.title("Histograma de Frequência dos Contextos de Afetos", fontsize=16)
plt.xlabel("Contexto/Ambiente", fontsize=12)
plt.ylabel("Frequência", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.plot();
```

```python
# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(sorted_labels, sorted_values, marker='o', linestyle='-', color='purple')
plt.title(title1, fontsize=16)
plt.xlabel("Contexto/Ambiente", fontsize=12)
plt.ylabel("Frequência", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.grid(True)
plt.tight_layout()
plt.plot();
```

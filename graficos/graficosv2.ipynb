```python
import matplotlib.pyplot as plt
```

# Bloco 1

### Gráfico 1 -
Contextualização dos significados dos empréstimos africanos encontrados no crioulo guineense

```python
title1 = "Contextualização dos significados dos empréstimos africanos encontrados no crioulo guineense"
collumn_type1 = [
    "Ambiente religioso",
    "Ambiente familiar",
    "Tratamento social",
    "Acolhimento"
]
collumn_values1 = [3, 3, 5, 1]
```

```python
# Create the line chart
plt.figure(figsize=(10, 6))
plt.pie(collumn_values1, labels=collumn_type1, autopct='%1.1f%%', colors=['purple', 'blue'])
plt.title(title1, fontsize=16)
plt.plot();
```

### Gráfico 2 - 
Contexto do empréstimo africano no crioulo guineense: é ligado ao afeto ou desafeto?

```python
# Data
labels2 = ['Afeto', 'Desafeto']
palavras2 = [9, 3]
expressoes2 = [3, 1]

# Create the plot with specified colors
plt.plot(labels2, palavras2, marker='o', linestyle='-', label='Palavras', color='blue')
plt.plot(labels2, expressoes2, marker='o', linestyle='-', label='Expressões', color='purple')

# Add titles and labels
plt.title('Contexto do empréstimo africano no crioulo guineense: é ligado ao afeto ou desafeto?')
plt.xlabel('Sentimento')
plt.ylabel('Contagem')
plt.legend()
plt.grid(True)
```
